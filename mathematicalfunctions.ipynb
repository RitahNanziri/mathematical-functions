# Linear Equations
This example represents the linear function 
#f(x) = 2 x + 3
#f(x)=2 x+3. You can customize the slope and y_intercept values to create different linear functions


Example One 

```python
def linear_function(x, slope, y_intercept):
    """
    Calculates the value of a linear function for a given x.
    
    Parameters:
    - x (float): The input value.
    - slope (float): The slope of the linear function.
    - y_intercept (float): The y-intercept of the linear function.
    
    Returns:
    float: The value of the linear function at the given x.
    """
    return slope * x + y_intercept

# Example usage:
slope = 2
y_intercept = 3

# Evaluate the linear function for x = 4
result = linear_function(4, slope, y_intercept)
print(f'The value of the linear function at x = 4 is {result}')
#In this example, the linear_function function takes three parameters (x, slope, and y_intercept) and returns the value of the linear function at the specified x value.
```

INFINITY IN RELATION TO  LINEAR AND NON LINEAR FUNCTIONS

Example, using a linear equation 
y = 2x + 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define linear function
def linear_function(x):
    return 2* x + 3
#generating x values
x_values=np.linspace(-10,10,100)

#calculate corresponding y values for the linear functions
y_values=linear_function(x_values)

# Plot the linear function
plt.plot(x_values,y_values, label="y=2x+3")

#Add arrows to indicate the directions towards infinity
plt.arrow(10,linear_function(10),2,4, head_width=0.5, head_length=1,fc="red", ec="red",label="Approaching Infinity")
plt.arrow(-10,linear_function(-10),-2,-4, head_width=0.5, head_length=1,fc="blue", ec="blue",label="Approaching negative Infinity")

#set plot limits
plt.xlim(-12,12)
plt.ylim(-20,25)

#add labels and legends
plt.xlabel("x")
plt.ylabel("y")
plt.title("linear function approaching infinity")
plt.legend()

#show the plot
plt.grid(True)

plt.show()
```

Non linear equations
